```python
# importing the libraries
import pandas as pd
from sklearn import linear_model,model_selection,metrics
import math
```

```python
# load the data from csv
house_train=pd.read_csv('F:/DataSciencePractise/MachineLearningLab/datasets/HomePrice/train.csv')
house_test=pd.read_csv('F:/DataSciencePractise/MachineLearningLab/datasets/HomePrice/test.csv')
```

```python
house_train.head()
```

```python
house_test.head()
```

```python
x_dims=['LotArea']
y_dims=['SalePrice']
```

```python
X_train=house_train[x_dims]
y_train=house_train[y_dims]
X_test=house_test[x_dims]
```

```python
lr=linear_model.Lasso()
lr.fit(X_train,y_train)

print("Coef : ",lr.coef_)
print("Intercept : ",lr.intercept_)
```

```python
# estimators will have scoring funciton
# Resubstitution strategy. Score is calculated using explained variance
lr.score(X_train,y_train)
y_predict=lr.predict(X_train)
house_test['SalePrice']=lr.predict(X_test)
```

```python
metrics.explained_variance_score(y_train,y_predict) # this is called train score
```

```python
meansqr=metrics.mean_squared_error(y_train,y_predict) # this is called train error
math.sqrt(meansqr)
```

```python
house_test[['Id','SalePrice']].to_csv('laso_lotarea_submission.csv',index=False)
```
